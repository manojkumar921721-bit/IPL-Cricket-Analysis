```python
import pandas as pd
import numpy as np
```

```python
matches = pd.read_csv("../data/raw/matches.csv")
deliveries = pd.read_csv("../data/raw/deliveries.csv")
```

```python
matches_df = matches.copy()
deliveries_df = deliveries.copy()
```

```python
matches_df.head()
```

```python
matches_df.tail()
```

```python
matches_df.shape
```

```python
deliveries_df.shape
```

```python
matches_df.columns
```

```python
matches_df.columns = matches_df.columns.str.lower()
deliveries_df.columns = deliveries_df.columns.str.lower()
```

```python
matches_df.columns
```

```python
matches_df.info()
```

```python
matches_df.isnull().sum()
```

```python
deliveries_df.isnull().sum()
```

```python
(matches_df.isnull().sum() / len(matches_df)) * 100
```

```python
matches_df["city"] = matches_df["city"].fillna("Unknown")
```

```python
matches_df["winner"] = matches_df["winner"].fillna("No Result")
```

```python
matches_df["player_of_match"] = matches_df["player_of_match"].fillna("Not Awarded")
```

```python
matches_df["umpire3"] = matches_df["umpire3"].fillna("Not Available")
```

```python
matches_df.isnull().sum()
```

```python
matches_df.duplicated().sum()
```

```python
deliveries_df.duplicated().sum()
```

```python
matches_df = matches_df.drop_duplicates()
deliveries_df = deliveries_df.drop_duplicates()
```

```python
matches_df.dtypes
```

```python
matches_df["date"] = pd.to_datetime(matches_df["date"], dayfirst=True)
```

```python
matches_df["date"] = pd.to_datetime(matches_df["date"])
```

```python
matches_df.dtypes
```

```python
matches_df["year"] = matches_df["date"].dt.year
matches_df["month"] = matches_df["date"].dt.month
matches_df["day"] = matches_df["date"].dt.day
```

```python
matches_df.head()
```

```python
matches_df["team1"].unique()
```

```python
matches_df["team1"].value_counts()
```

```python
team_mapping = {
    "Delhi Daredevils": "Delhi Capitals",
    "Rising Pune Supergiant": "Rising Pune Supergiants"
}
```

```python
team_columns = ["team1", "team2", "winner", "toss_winner"]

for col in team_columns:
    matches_df[col] = matches_df[col].replace(team_mapping)
```

```python
matches_df.describe()
```

```python
matches_df.to_csv("../data/processed/cleaned_matches.csv", index=False)

deliveries_df.to_csv("../data/processed/cleaned_deliveries.csv", index=False)
```
